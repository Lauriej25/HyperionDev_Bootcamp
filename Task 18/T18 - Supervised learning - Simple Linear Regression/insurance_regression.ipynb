```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
# Load "insurance.csv"
data = pd.read_csv("insurance.csv")

# Check the first few rows to identify relevant columns
data.head() 
```

```python
# Plot the data to show Age vs Charges, add a grid and colour to read easier 
plt.scatter(data['age'], data['charges'], color='blue', alpha=0.5)
plt.title('Age vs Insurance Charges')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.grid(True)
plt.show()
```

By looking at the data it shows as age increases so does the insurance charge

```python
# Show the line of best fit using linear regression
model = LinearRegression()
X = data[['age']]
y = data['charges']
model.fit(X, y)

# Make predictions
predictions = model.predict(X)

# Plot the best-fit line
plt.scatter(X, y, color='blue', alpha=0.5)
plt.plot(X, predictions, color='m')  # Plot the regression line
plt.title('Age vs Insurance Charges (with Best-Fit Line)')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.grid(True)
plt.show()
```

```python
# Ask user to input a age to predict the Insurance charge
age = int(input("Enter a age to predict the Insurance charge: "))
age = int(age)

# Define the input data
unk_x = pd.DataFrame([[age]], columns=['age'])

# Predict using the model
predicted_charge = model.predict(unk_x)

# Print the result
print("Your predicted insurance charge is: ", predicted_charge)

```
